```python
#standing

Acc_data = pd.read_csv('ACC_standing.csv')
Temp_data = pd.read_csv('TEMPstanding.csv')
EDA_data = pd.read_csv('EDA_standing.csv')
HR_data = pd.read_csv('HR_standing.csv')

#acc
Acc_length = Acc_data.shape[0]

#create a row of length 'acc_length' which is evenly spaced by seconds_ints

num_list = []
count = 0
sec_count =0

while count < Acc_length:
    num_list.append(sec_count)
    sec_count += seconds_int
    count += 1
    
idx=0
Acc_data.insert(loc=idx, column='A', value=num_list)
Acc_data

#convert datetime

Acc_data['A'] = pd.to_datetime(Acc_data['A'], unit='s')

Acc_data.columns =['A','acc_x','acc_y','acc_z']
#renaming columns


#temp 
Temp_data.columns = ['temp']
Temp_data.head()
Temp_length = Temp_data.shape[0]
temp_seconds_int = 1/4 

num_list = []
count = 0
sec_count =0

while count < Temp_length:
    num_list.append(sec_count)
    sec_count += temp_seconds_int
    count += 1
    
    
idx=0
Temp_data.insert(loc=idx, column='A', value=num_list)

#convert datetime

Temp_data['A'] = pd.to_datetime(Temp_data['A'], unit='s')


#EDA
EDA_data.columns = ['electroderm']
EDA_data.head()
EDA_length = EDA_data.shape[0]
EDA_seconds_int = 1/4 

num_list = []
count = 0
sec_count =0

while count < EDA_length:
    num_list.append(sec_count)
    sec_count += EDA_seconds_int
    count += 1
    
    
idx=0
EDA_data.insert(loc=idx, column='A', value=num_list)


#convert datetime

EDA_data['A'] = pd.to_datetime(EDA_data['A'], unit='s')


#HR

HR_data.columns = ['HeartRate']
HR_data.head()
HR_length = HR_data.shape[0]
HR_seconds_int = 1

num_list = []
count = 0
sec_count =0

while count < HR_length:
    num_list.append(sec_count)
    sec_count += HR_seconds_int
    count += 1
    
    
idx=0
HR_data.insert(loc=idx, column='A', value=num_list)

#convert datetime

HR_data['A'] = pd.to_datetime(HR_data['A'], unit='s')
```
